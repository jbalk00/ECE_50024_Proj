```python
#import statements
import os
import random
import shutil

#setup
random.seed(15)
```

```python
#define parameters
root = "/Users/jbalk/Documents/ML_Data/FinalProject/"
fld = "Images/"

shot1 = "1_Shot/"
shot5 = "5_Shot/"

train = "LSTM_Training/"
n_train = 64
num_train_es = 80

test = "LSTM_Testing/"
n_test = 20
num_test_es = 25

val = "LSTM_Validation/"
n_val = 16
num_val_es = 20

n = n_train+n_test+n_val

dtrain = "D_Train/"
dtest = "D_Test/"
```

```python
#Get list of categories
cats = [" "] * len(os.listdir(root+fld))
c = 0
for file in os.listdir(root+fld):
    cats[c] = file
    c = c + 1

cats.remove(".DS_Store")
print(cats)
```

```python
#Create file structure

def mkmydir(fname):
    if not os.path.exists(fname):
        os.makedirs(fname)

mkmydir(root+shot1+train)
mkmydir(root+shot1+test)
mkmydir(root+shot1+val)

mkmydir(root+shot5+train)
mkmydir(root+shot5+test)
mkmydir(root+shot5+val)
```

```python
#create training directories for 1 shot learning

n_train_imgs = 1
n_test_imgs = 15

c = 0
for i in range(0, num_train_es):
    lpth = root+shot1+train+"T"+str(c)+"/"
    ltrain = lpth+dtrain
    ltest = lpth+dtest
    mkmydir(lpth)
    mkmydir(ltrain)
    mkmydir(ltest)
    loc_cat_inds = random.sample(range(0, n_train), 5)

    for cind in loc_cat_inds:
        lcat = cats[cind]
        pics = os.listdir(root+fld+lcat+"/")
        inds = random.sample(range(0, len(pics)), n_train_imgs+n_test_imgs)
        mkmydir(ltrain+lcat)
        mkmydir(ltest+lcat)
        for w in range(0, n_train_imgs):
            shutil.copy2(root+fld+lcat+"/"+pics[inds[w]], ltrain+lcat+"/"+pics[inds[w]])
        for w in range(n_train_imgs, n_test_imgs+n_train_imgs):
            shutil.copy2(root+fld+lcat+"/"+pics[inds[w]], ltest+lcat+"/"+pics[inds[w]])

    c = c+1 #increment counter
```

```python
#create testing directories for 1 shot learning

n_train_imgs = 1
n_test_imgs = 15

c = 0
for i in range(0, num_test_es):
    lpth = root+shot1+test+"T"+str(c)+"/"
    ltrain = lpth+dtrain
    ltest = lpth+dtest
    mkmydir(lpth)
    mkmydir(ltrain)
    mkmydir(ltest)
    loc_cat_inds = random.sample(range(n_train, n_train+n_test), 5)

    for cind in loc_cat_inds:
        lcat = cats[cind]
        pics = os.listdir(root+fld+lcat+"/")
        inds = random.sample(range(0, len(pics)), n_train_imgs+n_test_imgs)
        mkmydir(ltrain+lcat)
        mkmydir(ltest+lcat)
        for w in range(0, n_train_imgs):
            shutil.copy2(root+fld+lcat+"/"+pics[inds[w]], ltrain+lcat+"/"+pics[inds[w]])
        for w in range(n_train_imgs, n_test_imgs+n_train_imgs):
            shutil.copy2(root+fld+lcat+"/"+pics[inds[w]], ltest+lcat+"/"+pics[inds[w]])

    c = c+1 #increment counter
```
